# 1. Furniture
For Furniture, we have 4 sub-categories, Tables, Furnishings, Chairs and Bookcases. Lets see how they are performing.

```python
# Lets create a dataframe for all products that fall under furniture

condition = sales_df['Category'] == 'Furniture'
furniture_df = sales_df[condition]

furniture_df.head()
```

```python
# Now, lets see the sales, profits and quantity sold  f the different sub-category products.
furniture_df = furniture_df.groupby(['year_only', 'Sub_Category']).sum(numeric_only=True).reset_index()

def furnitureGraphs():
    # Set up the subplots (3 columns for Sales, Profit, and Quantity Sold)
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))

    # Define the metrics to plot
    metrics = ['Sales', 'Profit', 'Quantity']
    colors = ['skyblue', 'lightgreen', 'lightcoral', 'goldenrod']

    # Plot the bar graphs for each metric
    for idx, metric in enumerate(metrics):
        # Pivot data to have Sub_Category as columns and Year as rows for each metric
        pivot_data = furniture_df.pivot(index='year_only', columns='Sub_Category', values=metric)

        # Plot
        pivot_data.plot(kind='bar', ax=axes[idx], color=colors, width=0.8)
        axes[idx].set_title(f'{metric} Over the Years by Sub-Category')
        axes[idx].set_xlabel('year_only')
        axes[idx].set_ylabel(metric)
        axes[idx].legend([])
        axes[idx].tick_params(axis='x', rotation=0)

    # Adjust the layout for better spacing
    plt.tight_layout()

    # Add a common legend at the bottom of all three subplots
    fig.legend(pivot_data.columns, loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=4, prop={'size': 20})

# Show the plots
furnitureGraphs()
```

## Organize the Furniture Data

```python
# Group by years and sub-categories to aggregate the values of sales, Profit and Quantity
furnitureSales_df = furniture_df.groupby(['year_only', 'Sub_Category'])['Sales'].sum().unstack()

furnitureProfit_df = furniture_df.groupby(['year_only', 'Sub_Category'])['Profit'].sum().unstack()

furnitureQuantity_df = furniture_df.groupby(['year_only', 'Sub_Category'])['Quantity'].sum().unstack()


# Now that we have the YoY increases, we can check for the perfomances of each sub-category 
furnitureSales_growth = furnitureSales_df.pct_change() * 100  # Calculate YoY growth for sales
furnitureSales_growth = furnitureSales_growth.round(2)
furnitureSales_growth = furnitureSales_growth.map(lambda x: f"{x:.2f}".rstrip('0').rstrip('.'))

furnitureProfit_growth = furnitureProfit_df.pct_change() * 100  # Calculate YoY growth for profits
furnitureProfit_growth = furnitureProfit_growth.round(2)
furnitureProfit_growth = furnitureProfit_growth.map(lambda x: f"{x:.2f}".rstrip('0').rstrip('.'))

furnitureQuantity_growth = furnitureQuantity_df.pct_change() * 100  # Calculate YoY growth for quantity
furnitureQuantity_growth = furnitureQuantity_growth.round(2)
furnitureQuantity_growth = furnitureQuantity_growth.map(lambda x: f"{x:.2f}".rstrip('0').rstrip('.'))

print(furnitureSales_growth)
print(furnitureProfit_growth)
print(furnitureQuantity_growth)
```

Now that we have broken down the 3 different categories and have seen their performance based on the metrics of Sales, Profit and Quantity, lets see what kind of customers are buying these products. These may show some insights to why some products do better than others.

## Who are the customers?

In the data cleaning process, we have checked for duplicate row values and found none. However, on closer inspection through the dataset using microsoft excel, I have found that there are repeat customers that buy a range of different products. Thus, I think we should look into who these customers are to find out our consumer demographic. 

```python
# lets see how many unique customers there are first.
sales_df.head()

# Count distinct customer IDs
distinct_count = sales_df["Customer_ID"].nunique()

# Check the distinct Segment labels
segment_values = sales_df['Segment'].unique()
print(segment_values)

print(f"Number of distinct customer IDs: {distinct_count}")
```

Now we know how many distinct customers there are, lets see what kind of customer they are. Through excel filtering, I have found there are 3 different types of customers; Consumer, Corporate and Home Office. Lets see the distribution of the 3. I am making the assumption that "CONSUMER" is just a normal everyday buyer which does not represent a buyer for a company or work.

```python
# Drop duplicate CustomerIDs
unique_customers = sales_df.drop_duplicates(subset="Customer_ID")

# Count occurrences of each customer type
customerType_count = unique_customers["Segment"].value_counts()

print(customerType_count)

customerType_df = pd.DataFrame(customerType_count)
customerType_df.head()

# Create a pie chart
customerType_df["count"].plot(
    kind='pie',
    autopct='%1.1f%%', 
    startangle=90, 
    figsize=(8, 8), 
    title="Customer Distribution"
)
plt.ylabel("")  # Hide y-axis label
plt.show();
```

Now that we see that normal everyday people are the majority of buyers, lets see for each category, who are the main customers. 
IMPORTANT TO NOTE: A customer may buy more than one product and of each different category, therefore, I just want to check which types of customers are buying the different category of products.

```python
# Remove duplicate CustomerID entries for each Category
unique_purchases = sales_df.drop_duplicates(subset=["Customer_ID", "Category"])

# Count purchases by Segment and Category
segment_count = unique_purchases.groupby(["Category", "Segment"]).size().unstack()

print(segment_count)

# Create a pie chart for each category
categories = segment_count.index
for category in categories:
    plt.figure(figsize=(6, 6))
    segment_count.loc[category].plot(
        kind="pie",
        autopct='%1.1f%%',
        startangle=90,
        title=f"Buyer Demographics for {category}",
        labels=segment_count.columns,
        colors=['skyblue', 'orange', 'lightgreen']
    )
    plt.ylabel("")  # Hide the y-axis label
plt.show();
```

We can see the demographics for each Category is pretty similar, with the average consumer being the highest, followed by Corporate and then Home Office. However, we would have to go one more level further, to see if theres a specific group that leads to more profits. Buyers from corporate may buy more expensive products or consumers may just be casual buyers. So lets check this out.

## Sales, Profit and Quantity metrics across the regions

```python
## Now lets check the sales, profit and quantity to see if there are any outliers or more insights
## We have only checked the customer demographic, but, maybe a lower customer count in one area may have higher profits due to the products bought

# Group by Region and Category and sum up the metrics
region_category_df = sales_df.groupby(['Region', 'Category']).sum(numeric_only=True).reset_index()

# Create Pivot Tables for Sales, Profit, and Quantity Sold
sales_pivot = region_category_df.pivot(index='Region', columns='Category', values='Sales').fillna(0)
profit_pivot = region_category_df.pivot(index='Region', columns='Category', values='Profit').fillna(0)
quantity_pivot = region_category_df.pivot(index='Region', columns='Category', values='Quantity').fillna(0)

# Plotting the grouped bar charts
fig, axes = plt.subplots(3, 1, figsize=(12, 18))

# Sales Plot
sales_pivot.plot(kind='bar', ax=axes[0])
axes[0].set_title('Sales Distribution Across Regions')
axes[0].set_xlabel('Region')
axes[0].set_ylabel('Total Sales')
axes[0].legend(title='Category')

# Profit Plot
profit_pivot.plot(kind='bar', ax=axes[1])
axes[1].set_title('Profit Distribution Across Regions')
axes[1].set_xlabel('Region')
axes[1].set_ylabel('Total Profit')
axes[1].legend(title='Category')

# Quantity Sold Plot
quantity_pivot.plot(kind='bar', ax=axes[2])
axes[2].set_title('Quantity Sold Distribution Across Regions')
axes[2].set_xlabel('Region')
axes[2].set_ylabel('Quantity')
axes[2].legend(title='Category')

plt.tight_layout()
plt.show();
```

This is interesting. As checked before, the number of customers in the west has the highest amount. However, the profit generated from the West is not the highest. From the bar graphs above, we can see that the East Region has the highest amount of Profit. Customers from the East region may be buying more expensive products as compared to the West, accounting for the difference in profit generated.
